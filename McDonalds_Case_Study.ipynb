1. EXPLORING DATA






```python
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
scaled_df = scalar.fit_transform(df1)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=11)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2","PCA3","PCA4","PCA5","PCA6","PCA7","PCA8","PCA9","PCA10","PCA11"])
pca_df.head()
```

2. Extracting Segments
> (a) Using k-means



```python
#Customer segmentation - based on pyschographic segmentation

#For convinence renaming the category
df['Like']= df['Like'].replace({'I hate it!-5': '-5','I love it!+5':'+5'})
#Like
sns.catplot(x="Like", y="Age",data=df,
            orient="v", height=5, aspect=2, palette="Set2",kind="swarm")
plt.title('Likelyness of McDonald w.r.t Age')
plt.show()
```

```python
#Using k-means clustering analysis
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df1)
visualizer.show()
```

```python
Sum_of_Squared_distances = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans = kmeans.fit(scaled_df)
    Sum_of_Squared_distances.append(kmeans.inertia_)
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
pca_df_kmeans.head()
```

```python
#To see each cluster size
from collections import Counter
Counter(kmeans.labels_)
```

```python
plt.figure(figsize=(12,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','orange','blue'])
plt.title("Clustering using K-Means Algorithm")
plt.show()
```

```python
#Creating a new coloumn to divide the likes into good and bad ratings

Likes_Score = [
    (df['Like'] ==  '+3'),
    (df['Like'] ==  '+2'),

    (df['Like'] ==  '+1'),
    (df['Like'] ==  '+4'),
    (df['Like'] ==  'I love it!+5'),
    (df['Like'] ==  '0'),
    (df['Like'] ==  'I hate it!-5'),
    (df['Like'] ==  '-3'),
    (df['Like'] ==  '-4'),
    (df['Like'] ==  '-2'),
    (df['Like'] ==  '-1'),
    ]

# Creating the list of the values that we want to assign likes to
values = ['Good Rating','Good Rating','Good Rating','Good Rating','Good Rating',
         'Neutral Rating','Bad Rating','Bad Rating','Bad Rating','Bad Rating','Bad Rating']

df['Ratings'] = np.select(Likes_Score,values)
df.head()
```

```python
df['Ratings'].value_counts()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df1)
df['cluster_num'] = kmeans.labels_ #adding to df
print (kmeans.labels_) #Label assigned for each data point
print (kmeans.inertia_) #gives within-cluster sum of squares.
print(kmeans.n_iter_) #number of iterations that k-means algorithm runs to get a minimum within-cluster sum of squares
print(kmeans.cluster_centers_) #Location of the centroids on each cluster.
```
